```python
import os
import pandas as pd
```

```python
csv_files = os.listdir(os.path.join('data', 'csv_paradise_papers.2018-02-14'))
csv_files
```

## Get overview of data structure

```python
data = {}
for csv_file in csv_files:
    file_type = csv_file.split('.')[-2]
    
    data[file_type] = pd.read_csv(os.path.join('data', 'csv_paradise_papers.2018-02-14', csv_file), low_memory=False)
```

```python
paradise_data = {}

for file_type, file_data in data.items():
    paradise_data[file_type] = pd.Series(file_data.columns.values)

pd.DataFrame(paradise_data).head(20)
```

```python
for file_type, file_data in data.items():
    print(f"{file_type}: has {file_data.shape[0]} rows and {file_data.shape[1]} columns \n")
    print(file_data.dtypes)
    print("-" * 100)
```

## On careful consideration, we can see all node csv have `node_id` which is unique identity of that information and edge csv have `START_ID` and `END_ID` which tells about connection between 2 nodes.
